```python
count_df = pd.DataFrame(pred_cancer['diagnosis'].value_counts().reset_index())

# Increase plot size
sns.set(rc={"figure.figsize": (15, 4)})

# Plot the bar plot
ax = sns.barplot(data=count_df, x="index", y="diagnosis")

# Rotate the x axis
ax.set_xticklabels(ax.get_xticklabels())

# Label Title
plt.title('Diagnosis Plot')

plt.show();
```

```python
corr = pred_cancer.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show();
```

```python
sns.pairplot(pred_cancer, hue='diagnosis');
```
